```python
def solve_maze(maze):
    n, m = len(maze), len(maze[0])
    for i in range(n):
        for j in range(m):
            if maze[i][j] == 'L':
                x0, y0 = i, j
    ans = ''
    last_visited_at = 0
    visited = {(x0, y0)}
    rev = {'R': 'L', 'L': 'R', 'U': 'D', 'D': 'U'}
    def dfs(x, y):
        nonlocal ans, last_visited_at
        for dx, dy, c in [(0, 1, 'R'), (0, -1, 'L'), (1, 0, 'D'), (-1, 0, 'U')]:
            x1, y1 = x + dx, y + dy
            if 0 <= x1 < n and 0 <= y1 < m and maze[x1][y1] != '#' and (x1, y1) not in visited:
                visited.add((x1, y1))
                ans += c
                last_visited_at = len(ans)
                dfs(x1, y1)
                ans += rev[c]
    dfs(x0, y0)
    print(len(ans), last_visited_at)
    return ans[:last_visited_at]

def rle(s):
    ans = ''
    i = 0
    cnt = 0
    while i < len(s):
        j = i
        while j < len(s) and s[j] == s[i]:
            j += 1
        ans += s[i] + str(j - i)
        i = j
        cnt += 1
    print("here", cnt)
    return ans

def encode_sol(sol, task_i):
    code = 1
    for c in sol:
        code = code * 4 + "RULD".index(c)
    
    @lam
    def f(rec, n):
        return If(n == 1, f"solve lambdaman{task_i} ", cat(rec(n / 4), str2expr("RULD").tail(n % 4).head_(1)))
    
    expr = Yc(f, code)
    # assert sol == run_fast(str(expr))
    return str(expr)

def encode_sol2(sol, task_i):
    return "S" + interpreter.encode_string(f"solve lambdaman{task_i} {sol}")

print(len(encode_sol("", 5)))
# for i in range(4, 6):
#     with open(f"problems/lambdaman/raw_lambdaman{i}.txt", "rt") as f:
#         maze = run_fast(f.read())
#     maze = maze.split()

#     sol_maze = solve_maze(maze)
#     print(sol, len(rle(sol_maze)))
#     sol = encode_sol(sol_maze, i)
#     print(i, len(maze), len(sol))
#     print(comm(sol))
```

```python
import random

class Rng:
    def __init__(self, seed):
        self.seed = seed

    def next(self):
        ret = self.seed % 3
        self.seed = (self.seed * 48271) % (2**31 - 1)
        return ret

with open("problems/lambdaman/raw_lambdaman18.txt", "rt") as f:
    maze = run_fast(f.read())
    maze = maze.split()
    best_iters = 1000000000
    while True:
        unvisited = set()
        for i in range(len(maze)):
            for j in range(len(maze[0])):
                if maze[i][j] == 'L':
                    x, y = i, j
                if maze[i][j] == '.':
                    unvisited.add((i, j))
        iters = 0
        d = 0
        seed = random.randint(0, 2**31 - 1)
        rng = Rng(seed)
        while unvisited:
            iters += 1
            if iters > best_iters:
                break
            dx, dy = [(0, 1), (1, 0), (0, -1), (-1, 0)][d]
            x1, y1 = x + dx, y + dy
            if 0 <= x1 < len(maze) and 0 <= y1 < len(maze[0]) and maze[x1][y1] != '#':
                x, y = x1, y1
                if (x, y) in unvisited:
                    unvisited.remove((x, y))
            d = (d + rng.next() + 3) % 4
        if iters < best_iters:
            best_iters = iters
            print(seed, iters)
```
